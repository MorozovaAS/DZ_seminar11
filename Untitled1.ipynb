```python
# 1. Определить корни f(x) = sin(x)**2 - cos(x)**2
from sympy import *
x = Symbol('x')
f = sin(x)**2 - cos(x)**2
solve (f, x)
```

```python
# 4. Построить график
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
def f(x):
  return np.sin(x)**2 - np.cos(x)**2
x = np.arange(-10, 10, 0.1)
plt.plot(x, f(x))
plt.show()

```

```python
# 5. Вычислить вершину

diff_f = diff (f, x)
solve(diff_f, x)
```

```python
# 6. Определить промежутки, на котором f > 0

solve (f > 0, x)
```

```python
# 7. Определить промежутки, на котором f < 0

solve (f < 0, x)
```

```python
# 2. Найти интервалы, на которых функция возрастает

solve(diff_f > 0, x)
```

```python
# 3. Найти интервалы, на которых функция убывает

solve(diff_f < 0, x)
```
